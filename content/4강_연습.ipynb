## Extra. (+5점 : 문제 1-3에서 감점된 경우 보너스로 인정)

(출처 : https://pokemon.fandom.com/ko/wiki/타입_상성표)  

##### 아래 그림은 포켓몬의 타입에 따른 상성을 나타낸 표입니다. `type_chart` 데이터프레임은 이를 바탕으로 생성되었습니다.  
- 각 행의 인덱스는 공격 스킬의 타입을, 각 열의 인덱스는 방어 포켓몬의 타입을 나타냅니다.
- 데이터프레임의 각 원소는 해당 공격 스킬의 타입으로 공격을 했을 때, 방어 포켓몬의 타입에 따라 입는 데미지의 배율을 나타냅니다.
- 예를 들어 100 데미지의 공격 스킬에 대해, `type_chart`의 원소가 0.5인 경우는 50의 데미지를 입는다는 의미입니다.
- 원소의 값이 `NaN`인 경우는 배율이 없음, 즉 원래 데미지 1배를 그대로 입는다는 의미입니다.

##### 공격받는 포켓몬의 타입이 1개인 경우는 타입 상성표에 있는 그대로 계산하면 됩니다.
##### 다만, 타입이 2개인 포켓몬은 각 타입의 해당하는 배율을 서로 곱한 결과를 사용합니다.   
- 예를 들어 불꽃(`Fire`)/비행(`Flying`) 타입인 **리자몽**에게 바위(`Rock`) 타입의 기술을 사용하면 불꽃 타입과 비행 타입 각각 바위타입 스킬로부터 2배의 데미지를 받기 때문에 총 $2\times 2 = 4$배의 데미지를 받게 됩니다.
- 반대로 리자몽에게 땅(`Ground`) 타입의 기술을 사용하면 불꽃 타입은 2배, 비행 타입은 0배의 데미지를 받기 때문에 총 $2\times 0=0$배의 데미지, 즉 **아무런 데미지를 받지 않게 됩니다**.
